```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import matplotlib.pylab as plt
import numpy as np
from sklearn.decomposition import PCA

data= pd.read_csv("Computers.csv")

pca = PCA(n_components=2)
dimension_low = pca.fit_transform(data)
print(dimension_low)


kmeans = KMeans(n_clusters=3).fit(dimension_low)
centroids = kmeans.cluster_centers_

print(centroids)
plt.scatter(dimension_low[:, 0], dimension_low[:, 1], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='D', c='red', s=50)
plt.show
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
data= pd.read_csv("Computers.csv")
pca = PCA(n_components=2)
dimension_low = pca.fit_transform(data)

classification1 = pd.DataFrame(columns=['price', 'speed', 'hd', 'ram', 'screen', 'ads'])
classification2 = pd.DataFrame(columns=['price', 'speed', 'hd', 'ram', 'screen', 'ads'])
classification3 = pd.DataFrame(columns=['price', 'speed', 'hd', 'ram', 'screen', 'ads'])

index1 = 0
index2 = 0
index3 = 0

kmeans = KMeans(n_clusters=3).fit(dimension_low)
centroids = kmeans.cluster_centers_
print(centroids)

for i in range(len(kmeans.labels_)):
    if kmeans.labels_[i] == 0:
        classification1.loc[index1] = data.iloc[i]
        index1 += 1
    elif kmeans.labels_[i] == 1:
        classification2.loc[index2] = data.iloc[i]
        index2 += 1
    else:
        classification3.loc[index3] = data.iloc[i]
        index3 += 1

print(classification1.mean())
print(classification2.mean())
print(classification3.mean())
print(classification1.describe())
print(classification2.describe())
print(classification3.describe())
plt.plot(classification1.mean())
plt.show()
plt.plot(classification2.mean())
plt.show()
plt.plot(classification3.mean())
plt.show()
plt.plot(classification1.mean())
plt.plot(classification2.mean())
plt.plot(classification3.mean())
plt.show()
```
